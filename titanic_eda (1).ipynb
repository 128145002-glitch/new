```
# Columns identified for type conversion
columns_to_convert = [
    'Daily_Usage_Time_minutes',
    'Social_Media_Usage_minutes',
    'Gaming_Usage_minutes',
    'Productivity_Usage_minutes'
]

for col_name in columns_to_convert:
    if col_name in df.columns:
        df[col_name] = pd.to_numeric(df[col_name], errors='coerce')
        print(f"Converted '{col_name}' to numeric.")

# Re-check info after conversion
print("\n### DataFrame Info after type conversion ###")
df.info()

# Re-check missing values after conversion (might reveal new NaNs from coercion)
print("\n### Missing Values after type conversion (Count) ###")
missing_values_after_conversion = df.isnull().sum()
display(missing_values_after_conversion[missing_values_after_conversion > 0])

# For simplicity, we'll fill new NaNs with the mean for numerical columns if any appeared.
for col_name in columns_to_convert:
    if col_name in df.columns and df[col_name].isnull().any():
        mean_val = df[col_name].mean()
        df[col_name].fillna(mean_val, inplace=True)
        print(f"Filled NaN in '{col_name}' with mean: {mean_val:.2f}")

print("\n### Missing Values after filling NaNs (Count) ###")
display(df.isnull().sum()[df.isnull().sum() > 0])
```
